# Step5: Data quality and consistency checks

```python
# Check for missing values in the whole dataset

customer.isnull().sum()
```

```python
# view the observation with the missingvalue in the first_name column

customer_nan=customer[customer['first_name'].isnull()==True] # creating a sub-set dataframe containing only observation with missing value 
```

```python
customer_nan
```

```python
# Filtering out observation with missing value (creating a new dataframe only with non-missing value)

customer_clean = customer[customer['first_name'].isnull() == False]
```

```python
customer_clean
```

```python
# Finding duplicates 

customer_dups= customer_clean[customer_clean.duplicated()]
```

```python
customer_dups
```

```python
# Checking the shape of the new dataframe

customer_clean.shape
```

```python
# Check if the dataframe contains mixed type data

for col in customer_clean.columns.tolist():
  weird = (customer_clean[[col]].applymap(type) != customer_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customer_clean[weird]) > 0:
    print (col)
```
