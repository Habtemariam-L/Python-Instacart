# Deriving Columns with loc()

```python
# Creating a flag column that assigns 'loyalty' label

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#.  Step 4 create a loyalty flag using the transform() and loc() functions. First aggregate data with transform () function

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Assign flag values using conditions 

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head()
```

```python
# Print the frequency of the loyalty flag column

ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# Step 5 Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers 

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

On average regular and new customers spend more than loyal customers.

```python
# Step 6 Create a spending flag. First aggregate data with the transform() function 

ords_prods_merge['avg_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Assign flag values using conditions 

ords_prods_merge.loc[ords_prods_merge['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Print the frequency of the spending flag column

ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
# Step 7 Create an order frequency flag. First aggregate data with the transform functin

ords_prods_merge['ord_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Assign flag values using conditions 

ords_prods_merge.loc[ords_prods_merge['ord_freq'] > 20, 'order_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['ord_freq'] <= 20) & (ords_prods_merge['ord_freq'] > 10), 'order_flag'] = 'Regular customer'

ords_prods_merge.loc[ords_prods_merge['ord_freq'] <= 10, 'order_flag'] = 'Frequent customer'
```

```python
# Print the frequency of the spending flag column

ords_prods_merge['order_flag'].value_counts(dropna=False)
```
