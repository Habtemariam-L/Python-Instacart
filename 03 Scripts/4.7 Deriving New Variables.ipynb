# If-Statements with User-Defined Functions

```python
# Define a function called price_label and set the conditions to be aplied on the prices column in the the newly created df

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
# Apply the function created above. The following will create a new column called price_range and checks every row (axis=1) 

df['price_range']=df.apply(price_label, axis=1)
```

```python
# Print the frequency of the price_range column

df['price_range'].value_counts(dropna=False)
```

# If-Statements with the loc() Function

```python
# Using the built-in loc function condition-1

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Using the built-in loc function condition-2

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
 # Using the built-in loc function condition-3
    
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Using the built-in loc function condition-4

df['price_range_loc'].value_counts(dropna=False)
```

```python
# using the loc function on the entire dataframe. the loc function is faster than the user defined function therefore there wouldnt be a memory issue to apply it on the netire dataframe  


df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

# If-Statements with For-Loops

```python
# Creating an empty list called result then apply the loop to populate the empty list 
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# creating a new column and set it equal to result

df_ords_prods_merge['busiest_day']=result
```

```python
# Print the frequency of the newly created busiest_day column

df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

# Tasks

```python
# Step 2 Creat a new column showing busiest_days, slowest_days and regular_days. First step is to creat an empty list called result and populate it with a loop functin

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value==1:
    result.append("Busiest days")
  elif value == 4 or value==3:
    result.append("Slowest days")
  else:
    result.append("Regularly days")
```

```python
# Creating a new column and set it equal to result

df_ords_prods_merge['busiest_days']=result
```

```python
# Print the frequency of the newly created  busiest_days column

df_ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

The busiest days frequency is equal to the sum of "0" and "1". And the slowest_days frequency is equal to the sum of "3" and "4". The frequency of regular_days is the sum of the rest of the days.

```python
# Step 4  Creat a new column called busiest_period_of_the_day. For this first check the frequency of hour_at_time_of_order column to decide the conditions 

df_ords_prods_merge['hour_at_time_of_order'].value_counts(dropna=False)

```

```python
# Creating a new list called result .

result = []

for value in df_ords_prods_merge["hour_at_time_of_order"]:
  if value >= 8 and value <=19 :
    result.append("Most orders")
  elif value >=2 and value<= 5:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
# Creating a new column and set it equal to the above result list

df_ords_prods_merge['busiest_periods_of_day']=result
```

```python
# Check the frequency of the new column busiest_periods_of_day

df_ords_prods_merge['busiest_periods_of_day'].value_counts(dropna=False)
```
